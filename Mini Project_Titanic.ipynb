```python
def age_fill(data):
    Age=data[0]
    pc=data[1]
    
    if pd.isnull(Age):
        if pc==1:
            return 38
        elif pc==2:
            return 28
        else:
            return 24
    else:
        return Age
```

```python
data["Age"]=data[["Age","Pclass"]].apply(age_fill,axis=1)
```

```python
data.info()
print("As you can see that Null values are replaced by median With respect to Pclass")
```

```python
# Preparing training data requires the removal of some column like 
'''
1)Name
2)Cabin
3)Ticket
4)Passenger Id
Because,their don't contribute much,and each of them are unique,so cannot find correlation with target. 
'''

data=data.drop(labels=["PassengerId","Name","Ticket","Cabin"],axis=1)
```

```python
#Converting Embarked and Sex to the required format(0,1,2 like),since they are categorical in nature

data["Sex"]=data["Sex"].replace({"male":0,"female":1})
data["Embarked"]=data["Embarked"].replace({'S':0, 'C':0, 'Q':1})
```

```python
#Dropping null and their espective rows
data=data.dropna(axis=0)
```

```python
#Checking length of data
len(data)
```

```python
#Checking the column and their respective data types and checking whether null values are  available or not.

data.info()
```

```python
#dividing data to x and y
x=data.drop(["Survived"],axis=1)
y=data["Survived"]
```

```python
#Splitting data into train and test.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

### 1.Logistic Regression

```python
#Getting object of LogisticRegression
LogReg=LogisticRegression()
```

```python
#Fitting Logistic Regression
LogReg.fit(x_train,y_train)
```

```python
#Prediction for the split x_test data
y_pred=LogReg.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print("Accuracy Score thorugh Logistic Regression=",(95+48)/(95+48+10+25))
```
